```python
##################### HEALPY 1

#HEALPIX = Hierarchical Equal Area isoLatitude Pixelization of a sphere, ferramenta usada para leitura 'visual' de mapas do céu, separa a esfera em pixels com mesma área                 
#o healpix foi criado para processar dados sobre a CMB obtidos pelos BOOMERANG e WMAP, hoje em dia usado para mais coisas  


print('#########')
###  NSIDE é a resolução do mapa
NSIDE = 2 # NSIDE=2^n , só pode potencias de 2
print("Approximate resolution at NSIDE {} is {:.2} deg".format(NSIDE, hp.nside2resol(NSIDE, arcmin=True) / 60))
NPIX = hp.nside2npix(NSIDE)# numero de pixels NPIX = 12*NSIDE^2
print('NPIX =',NPIX)
# UM MAPA EH UM ARRAY COM NUMEROS, UMA MATRIZ COM NUMEROS, np.array
print('#########')


### ORDENAÇAO dos pixels no mapa, há dois tipos, RING e NEST
mapa = np.arange(NPIX) #vetor 1, 2, 3, ...  de 0 ate NPIX-1
#criando vetor vec para visualização
vec = 0*mapa #multipliquei todos os valores por 0, assim teremos somente 0 de mesmo tamanho de m.
vec = vec.astype(bool) #transformei em um booleano. 0=False e 1=True.
pixel = 5 ############################### MUDAR AQUI
vec[pixel]=True
### loop para visualização dos mapas lado a lado
#enxergando as posições RING dos pixels, em amarelo 
mapas=[mapa, vec] #vetor com os mapas
titulos=["Mollview image RING","posiçao dos pixeis no mapa"] #vetor com os nomes
plt.figure(figsize = (14,4))
for i in range(2):
    hp.mollview(mapas[i],sub = (1,2,i+1), title = titulos[i], nest=False)
#enxergando as posições NEST dos pixels, em amarelo
titulos=["Mollview image NEST","posiçao dos pixeis no mapa"] #vetor com os nomes
plt.figure(figsize = (14,4))
for i in range(2):
    hp.mollview(mapas[i],sub = (1,2,i+1), title = titulos[i], nest=True)

```
